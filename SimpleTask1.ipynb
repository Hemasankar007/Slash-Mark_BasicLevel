```python
import random

class Task:
    def __init__(self, description, priority):
        self.description = description
        self.priority = priority

# Initialize an empty task list
tasks = []

# Load pre-existing tasks from a file (if any)
try:
    with open('tasks.txt', 'r') as file:
        for line in file:
            description, priority = line.strip().split(',')
            tasks.append(Task(description, priority))
except FileNotFoundError:
    pass

# Function to save tasks to a file
def save_tasks():
    with open('/content/tasks.csv', 'w') as file:
        for task in tasks:
            file.write(f"{task.description},{task.priority}\n")

# Function to add a task to the list
def add_task(description, priority):
    global tasks  # Declare tasks as a global variable
    tasks.append(Task(description, priority))
    save_tasks()
    print("Task added successfully.")

# Function to remove a task by description
def remove_task(description):
    global tasks
    tasks = [task for task in tasks if task.description != description]
    save_tasks()
    print("Task removed successfully.")

# Function to list all tasks
def list_tasks():
    if not tasks:
        print("No tasks available.")
    else:
        for task in tasks:
            print(f"Description: {task.description} - Priority: {task.priority}")

# Function to recommend a task based on a rule-based approach
def recommend_task():
    if not tasks:
        print("No tasks available for recommendations.")
        return

    # Sort tasks by priority (High > Medium > Low)
    sorted_tasks = sorted(tasks, key=lambda x: ['Low', 'Medium', 'High'].index(x.priority))

    # Choose the highest priority task
    recommended_task = sorted_tasks[0]

    print(f"Recommended task: {recommended_task.description} - Priority: {recommended_task.priority}")

# Main menu
while True:
    print("\nTask Management App")
    print("1. Add Task")
    print("2. Remove Task")
    print("3. List Tasks")
    print("4. Recommend Task")
    print("5. Exit")

    choice = input("Select an option: ")

    if choice == "1":
        description = input("Enter task description: ")
        priority = input("Enter task priority (Low/Medium/High): ").capitalize()
        add_task(description, priority)

    elif choice == "2":
        description = input("Enter task description to remove: ")
        remove_task(description)

    elif choice == "3":
        list_tasks()

    elif choice == "4":
        recommend_task()

    elif choice == "5":
        print("Goodbye!")
        break

    else:
        print("Invalid option. Please select a valid option.")
```
